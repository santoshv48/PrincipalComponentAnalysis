```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv('wine.csv')
```

```python
wine.head()
```

```python
winetype = wine['Type']
winetype
```

```python
wine = wine.iloc[:,1:]
wine
```

```python
wine.shape
```

```python
wine.info()
```

```python
#converting data to numpy array
wine_array = wine.values
wine_array
```

```python
#normalizing the numerical data
wine_norm = scale(wine_array)
wine_norm
```

# Pca implementation

```python
# applying pca fit transform to data set
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
#pca component matrix or covariance matrix
pca.components_
```

```python
# the amount of variance that each pca has
var = pca.explained_variance_ratio_
var
```

```python
#cummulative variance of each variance
var = np.cumsum(np.round(var,decimals = 4)*100)
var
```

```python
plt.plot(var,color='blue')
```

```python
# Final Dataframe
final_df=pd.concat([winetype,pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#visualization of pcas
import seaborn as sns
fig = plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df,x = 'PC1',y='PC2',hue='Type')
```

# Cheecking with other clustering alogarithms

 Hierarchical algorithms

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# as we already have normalized data,create dendogram
plt.figure(figsize=(10,8))
dendogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#Create Clusters
hclusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage = 'ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns = ['hcluster'])
y['hcluster'].value_counts()
```

```python
#adding clusters to dataset
wine2 = wine.copy()
wine2['hcluster'] = hclusters.labels_
wine2
```

# Kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
    
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
wine2['clusters3id']=clusters3.labels_
wine2
```
